- Learning rate, Œ±, is the size of step taken by each gradient.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# y = (x+3)**2
# slope = 2*(x+3)

slope = lambda x : 2*(x+3)
```

```python
x = 2
learning_rate = 0.01
precision = 0.000001
max_it = 500

prev_step_size = 1
it = 0
```

```python
while prev_step_size > precision and it < max_it:
    prev_x = x
    x = x - learning_rate*slope(x)
    
    prev_step_size = abs(x-prev_x)
    
    it = it+1
    print ("it: ", it, "-", x)
```
